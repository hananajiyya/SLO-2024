```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
from os import path
path = '/content/drive/MyDrive/SLO/data0-SZL.csv'
data = pd.read_csv(path)
```

```python
data.head()
```

```python

# Buat kolom baru untuk menyimpan statistik
data['Rata-rata'] = data.mean(axis=1)
data['Standar Deviasi'] = data.std(axis=1)
data['Skewness'] = data.skew(axis=1)
data['Median'] = data.median(axis=1)
data['Energi'] = np.sum(data**2, axis=1)
data['Min'] = data.min(axis=1)
data['Max'] = data.max(axis=1)

# Cetak dataframe hasil
print(data)

```

```python
data.head()
```

```python
# Pilih 6 kolom secara bebas
kolom_plot = ["953", "957", "989", "942", "1211", "952.1"]

plt.figure(figsize=(10, 6))

for i, kolom in enumerate(kolom_plot):
    plt.subplot(2, 3, i + 1)
    plt.plot(data[kolom])
    plt.title(kolom)

plt.show()
```

```python
colors = ['red', 'blue', 'green', 'orange', 'purple', 'brown', 'pink', 'gray', 'cyan', 'magenta']
markers = ['*', 'o', 'v', '^', 's', 'D', 'X', 'P', '>', '<']
marker_index = 0
for i in range(0, 30, 10):
    selected_columns = data.columns[i:i+10]
    color = colors[i // 10]
    marker = markers[marker_index]
    marker_index += 1
    for column in selected_columns:
        plt.scatter(data['Rata-rata'], data['Standar Deviasi'], color=color, marker=marker, label=f'Columns {i+1}-{i+10}')
```
